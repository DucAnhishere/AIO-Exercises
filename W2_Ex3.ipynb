```python
file_path = '/content/P1_data.txt'
if os.path.exists(file_path):
    with open(file_path, 'r') as f:
        content = f.read()
        print(content)
else:
    print(f"File {file_path} không tồn tại")
```

```python
def count_words(text):
    text = text.lower()
    word_count = {}
    for word in text.split():
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return word_count
```
